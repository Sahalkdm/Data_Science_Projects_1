```python
#X_train = df_train[feats]

'''
# Display the shapes of X and Y matrices
print("X_train.shape:", X_train.shape, "\nY_train.shape:", Y_train.shape)
Q1 = df_train.quantile(0.25)
Q3 = df_train.quantile(0.75)
IQR = Q3 - Q1

# Remove outliers
X_train = X_train[~((X_train < (Q1 - 1.5 * IQR)) |(X_train > (Q3 + 1.5 * IQR))).any(axis=1)]
# Pre-processing functions to fill NaN values with zero'''
from scipy.stats import zscore
from sklearn.preprocessing import StandardScaler
    
def process_X(X_inp):
    """
    Function to process the feature matrix X by replacing NaN values with zero.
    """
    df=pd.DataFrame(X_inp)

    for col in df.columns:
        print(col)
        print('initial')
        print(df[col].skew())
        print(df[col].kurtosis())
        df[col].fillna(0, inplace=True)
        '''upper_limit = df[col].quantile(0.99)
        lower_limit = df[col].quantile(0.01)'''
        '''if ((df[col] < lower_bound[col]) | (df[col] > upper_bound[col])).any():
            df[col] = df[col].mask(df[col] < lower_bound[col], lower_bound[col])
            df[col] = df[col].mask(df[col] > upper_bound[col], upper_bound[col])'''
        '''df[col] = np.where(df[col] >= upper_limit,
        upper_limit,
        np.where(df[col] <= lower_limit,
        lower_limit,
        df[col]))'''
        if (abs(df[col].skew())>2 or abs(df[col].kurtosis())>=3) and min(df[col]) > -0.02:
            df[col] = np.log(df[col] + 1)
        if (abs(df[col].skew())>2 or abs(df[col].kurtosis())>=3) and min(df[col]) > -0.02:
            df[col] = np.log(df[col] + 1)
        if (abs(df[col].skew())>2 or abs(df[col].kurtosis())>=3) and min(df[col]) > -0.02:
            df[col] = np.log(df[col] + 1)
        print('afterlog')
        print(df[col].skew())
        print(df[col].kurtosis())
        df[col] = zscore(df[col])
        print('afterZtr')
        print(df[col].skew())
        print(df[col].kurtosis())
    #out_X_inp = np.nan_to_num(X_inp)
    return df

def process_Y(Y_inp):
    """
    Function to process the target vector Y by replacing NaN values with zero.
    """
    out_Y_inp = np.nan_to_num(Y_inp)
    return out_Y_inp

X_train=process_X(X_train)
#scaler = StandardScaler()
#X_train = scaler.fit_transform(X_train)
#X_train = df_merge_imp[combined_df['is_train'] == 1]
#X_test = df_merge_imp[combined_df['is_train'] == 0]
#X_test = process_X(df_test_imputed)
#Y_train = df_train['target'].to_numpy()
Y_train = process_Y(Y_train)
```
